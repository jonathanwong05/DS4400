```python
import pandas as pd
import numpy as np
```

```python
# Load dataset
df = pd.read_csv('kc_house_data.csv')

# Drop unnecsessary columns
df = df.drop(columns=['id', 'date', 'zipcode'])

# Separate features and target variable
target = 'price'
features = [col for col in df.columns if col != target]
```

```python
# Part 1

# Compute statistics for each feature
stats = pd.DataFrame({
    'mean': df[features].mean(),
    'min': df[features].min(),
    'max': df[features].max(),
    'variance': df[features].var()
})

print("\nStatistics for each feature:")
print(stats.round(3))

# Lowest and highest average
lowest_avg_feature = stats['mean'].idxmin()
highest_avg_feature = stats['mean'].idxmax()

print(f"Lowest average: {lowest_avg_feature} = {stats.loc[lowest_avg_feature, 'mean']:.3f}")
print(f"Highest average: {highest_avg_feature} = {stats.loc[highest_avg_feature, 'mean']:.3f}")

# Lowest and highest variance
lowest_var = stats['variance'].idxmin()
highest_var = stats['variance'].idxmax()

print(f"Lowest variance: {lowest_var} = {stats.loc[lowest_var, 'variance']:.3f}")
print(f"Highest variance: {highest_var} = {stats.loc[highest_var, 'variance']:.3f}")
```

```python
# Part 2

# Compute correlation of each feature with price
correlations = df[features].corrwith(df[target])
corr_table = pd.DataFrame({
    'correlation_with_price': correlations
})

print("\nCorrelation of each feature with price:")
print(corr_table.round(3))

# Positively correlated features
positive_corr = corr_table[corr_table['correlation_with_price'] > 0]
print("\nPositively correlated features with price:")
print(positive_corr.round(3))

# Highest positive correlation
highest_pos_corr_feature = corr_table['correlation_with_price'].idxmax()
highest_pos_corr_value = corr_table['correlation_with_price'].max()
print(f"\nHighest positive correlation: {highest_pos_corr_feature} = {highest_pos_corr_value:.3f}")
```

```python
# Part 3

# Negatively correlated features
negative_corr = corr_table[corr_table['correlation_with_price'] < 0]

if len(negative_corr) > 0:
    print("\nFeatures with negative correlation:")
    print(negative_corr.round(3))
else:
    print("\nNo features have negative correlation found.")
```
