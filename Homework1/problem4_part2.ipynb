```python
import numpy as np
```

```python
np.random.seed(42)
```

```python
matrices = [np.random.randint(-10, 11, size=(3, 3)) for _ in range(3)]

for i, M in enumerate(matrices):
    print(f"{'='*50}")
    print(f"MATRIX {i+1}")
    print(f"{'='*50}")
    print(f"\nM{i+1} =")
    print(M)
    
    # Compute determinant to check if inverse exists
    det = np.linalg.det(M)
    print(f"\nDeterminant: {det:.4f}")
    
    if np.abs(det) > 1e-10:  # Matrix is invertible
        # Compute inverse
        M_inv = np.linalg.inv(M)
        print(f"\nM{i+1}^(-1) =")
        print(np.round(M_inv, 4))
        
        # Verification: M @ M_inv should equal identity matrix
        product = M @ M_inv
        print(f"\nVerification: M{i+1} @ M{i+1}^(-1) =")
        print(np.round(product, 10))
        
        # Confirm result is identity
        identity = np.eye(3)
        is_identity = np.allclose(product, identity)
        print(f"\nEquals identity matrix? {is_identity} ✓" if is_identity else f"\nEquals identity matrix? {is_identity} ✗")
    else:
        print("\nMatrix is SINGULAR - no inverse exists")
    
    print()
```
